```
#Create the data file
import pandas as pd
data = pd.read_csv('./keras-yolo3/model_data/stage_2_train_labels_resized.csv')
data.head()
```

```
img_path = './images/'
with open('./keras-yolo3/model_data/traindata.txt','w') as f:
  for row in data.values:
    img_fp = os.path.join(img_path, "{}.jpg".format(row[0]))
    if os.path.exists(img_fp):
      line = "{} {},{},{},{},{}\n".format(img_fp,int(row[1]),int(row[2]),int(row[3]),int(row[4]),int(row[5]))
      #line = "{}\n".format(img_fp)
    f.write(line)
f.close()
```
